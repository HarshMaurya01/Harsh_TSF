```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(rescale=1/ 255,
                            rotation_range=.20,
                            height_shift_range=.20,
                            width_shift_range=0.2,
                            horizontal_flip=True)
```

```python
train_data = datagen.flow_from_directory(train_path, target_size=(256,256))
test_data = datagen.flow_from_directory(test_path,target_size=(256,256))
```

```python
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten
```

```python
model = Sequential()
```

```python
model.add(Conv2D(32, (3,3),activation='relu',input_shape=(256,256,1)))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(32, (3,3),activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(64, (3,3),activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(64, (3,3),activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(128, (3,3),activation='relu'))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))

          
```

```python
model.summary()
```

```python
model.compile(optimizer='sgd',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(train_data, epochs=200)
```

```python
cnn = tf.keras.models.Sequential()

cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=(256, 256, 1)))

cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))

cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=(256, 256, 1)))

cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))

cnn.add(tf.keras.layers.Flatten())

cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))

cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

cnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
cnn.fit_generator(train_data, validation_data=test_data, epochs=25)
```

```python
from keras.applications.vgg19 import VGG19
from keras.applications.vgg19 import preprocess_input
```

```python
vgg19 = VGG19(weights='imagenet',include_top=False,input_shape=(256,256,3))
```
